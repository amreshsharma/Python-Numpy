```
#import numpy library

import numpy as np

#print numpy version

print(np.__version__)
```

```
# null vector of size 5

a = np.zeros(5)

print(a)

#change value of 2nd to 1
a[3] = 1

print('New elements of a after change'
     )
print(a)
```

```
#Create a vector with values ranging from 5 to 19

a = np.arange(5,20)

print(a)
```

```
#How to generate five random numbers

a = np.random.normal(size=5)

print(a)

# generate 4 random integers between 10 and 20.

a_4 = np.random.randint(low=10, high=20, size=4)

print('4 random numbers between 10 and 20')

print(a_4)
```

```
# How to find min, max values

a = np.array([5, 0, 3, 3, 7, 9])

print(a)

# min value

a_min = min(a)
print('Minimum :  ',a_min)

# max value

a_max = max(a)
print('Maximum :  ',a_max)
```



**Numpy Array Indexing**

```
a = np.array([5, 0, 3, 3, 7, 9])

print(a)

print('First element: ',a[0])      # indexing starts from 0

print('third element: ',a[2])
```

```
# To index from the end of the array, you can use negative indices:

print( a[-1] )
```

```
#In a multi-dimensional array, items can be accessed using a comma-separated tuple of indices:

a2 = np.array([[3, 5, 2, 4],[7, 6, 8, 8],[1, 6, 7, 7]])

print(a2)

# get first item

print('at 0,0 Position')

print(a2[0,0])

print('at 0,2 Position')

print(a2[0,2])
```

Array Values can be modified using index notation:

```
a = np.array([5, 0, 3, 3, 7, 9])

print(a)

# change value of first index 

a[0] = 22

print('New array')

print(a)
```

**Array Slicing**

Syntex : [start:stop:step]

```
a = np.arange(10)

print(a)

# first four elements

print( a[:4] )

# elements after index 4

print( a[4:] )


# every other element, starting at index 1

print( a[1::2]  )
```
